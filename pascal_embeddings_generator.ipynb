# Transfer Learning on PASCAL VOC 5i

```python
from pascal_embeddings_generator import main, plot_stats
import numpy as np
import pandas as pd
```

## Train Model

```python
train_kwargs = {
    "image_shape":(3, 384, 512),
    "epochs":30, 
    "freq":5, 
    "bs":15,
    "experiment_number": 1,
    "generate_embeddings": False,
    "model": "mobilenet_v2"
}

training_stats, iou_per_class_list, model = main(**train_kwargs)
```

## Plot Stats

```python
plot_stats(pd.DataFrame(training_stats), "val loss")
```

```python
from matplotlib import pyplot as plt
sorted_ious = [k for k,v in sorted(iou_per_class_list[-1].items(), key=lambda x:x[1], reverse=True)]
class_ious = pd.DataFrame(iou_per_class_list)
plt.figure(figsize=(12, 6))
for class_ in sorted_ious:
    plt.plot(class_ious.loc[:,class_].values, label=class_)    
plt.xlabel("Epochs")
plt.ylabel("Mean IOU per Class")
plt.title("Mean IOUs per Class vs Epochs")
plt.legend()
plt.grid()
```
